````python
import pandas

def load(path: str) -> pandas.core.frame.DataFrame:
    try:
        return pandas.read_csv(path)
    except FileNotFoundError as msg:
        print(msg)
    return pandas.core.frame.DataFrame()

print(load("./life_expectancy_years.csv"))
````

# Draw my country

***

Create a program that calls the load function from the previous exercise, loads the file<br>
life_expectancy_years.csv, and displays the country information of your campus. Your<br>
graph must have a title and a legend for each axis.<br>
For example, for the 42 campuses in France we will have this result<br>

***

````python
import matplotlib.pyplot as plt


def ft_plot(country: str, country_data: list[str]) -> None:
    plt.plot([_ for _ in range(1800, 2100)], country_data[1:])
    plt.title(f'Life Expectancy in {country}')
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy')
    plt.xlim(1800, 2100)
    plt.ylim(0, 100)


def get_country_data(country: str) -> list[str]:
    """Show country data
    
    Details:
    Plot the graph of the country life expectancy along the time
    """
    ds = load("./life_expectancy_years.csv")
    return ds[ds["country"] == country].values.flatten()[1:]

france_life_expectancy = get_country_data("France")
ft_plot("France", france_life_expectancy)
plt.show()
````

# Compare my country

***

Create a program that calls the load function from the first exercise, loads the file<br>
population_total.csv, and displays the country information of your campus versus other<br>
country of your choice. Your graph must have a title, a legend for each axis and a<br>
legend for each graph.<br>

You must display the years from 1800 to 2050.<br>

***

````python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

def text_to_num(text):
    """text_to_num

    Details:
    Convert string number to decimal.
    Accepts suffixes: [K, M, B, T]
    """
    d = {
        "K": 3,
        "M": 6,
        "B": 9,
        "T": 12
    }
    if text[-1].upper() in d:
        num, magnitude = text[:-1], text[-1].upper()
        return float(num) * 10 ** d[magnitude]
    else:
        return float(text)


def plot_countries_between(countries: list[str], start_year, end_year) -> None:
    """Get country data
    
    Details:
    Get a country data between two dates both dates are included
    """
    ds = load("./population_total.csv")
    ds_start_year = int(ds.columns[1])
    ds_end_year = int(ds.columns[-1])
    if start_year < ds_start_year or end_year > ds_end_year:
        print("Warning: The dates you gave are not available in the dataset")
    if start_year > end_year:
        print("Warning: Start is bigger than end")
    start_index =  start_year - ds_start_year + 1
    end_index = end_year - ds_start_year + 2
    for country in countries:
        if (ds[ds["country"] == country].empty):
            next
        country_pop = ds[ds["country"] == country].values.flatten()[start_index:end_index]
        country_pop = [text_to_num(_) for _ in country_pop]
        plt.plot([_ for _ in range(start_year, end_year+1)], country_pop, label=country)


def formatter(x, pos):
    """
    Convert the tick valie to a string with a suffix.

    '': none
    'K': thousands
    'M': millions
    'B': billions
    'T': trillions
    """
    suffixes = ['', 'K', 'M', 'B', 'T']
    order = 0
    while x >= 1000 and order < len(suffixes) - 1:
        order += 1
        x /= 1000.0
    return f"{x:.0f}{suffixes[order]}"

countries = ["Portugal", "France", "Mexico", "Belgium"]
start_year = 1800
end_year = 2050
plot_countries_between(countries, start_year, end_year)

plt.title(f'Population Projections')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Population')
plt.gca().yaxis.set_major_formatter(FuncFormatter(formatter))
plt.show()
````

````python
def text_to_num(text):
    """text_to_num

    Details:
    Convert string number to decimal.
    Accepts suffixes: [K, M, B, T]
    """
    d = {
        "K": 3,
        "M": 6,
        "B": 9,
        "T": 12
    }
    if type(text) == str and text[-1].upper() in d:
        num, magnitude = text[:-1], text[-1].upper()
        return float(num) * 10 ** d[magnitude]
    else:
        return float(text)


def plot_income_against_life_expectancy(countries, year):
    ds_income = load("./income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
    ds_expectancy = load("./life_expectancy_years.csv")
    fig, ax = plt.subplots()
    ax.set_xscale('log')
    if len(ds_income.get(str(year), [])) == 0:
        print("Warning, year not founded in the data set")
        return
    for country in countries:
        income = ds_income[ds_income["country"] == country][str(year)].values
        expectancy = ds_expectancy[ds_expectancy["country"] == country][str(year)].values
        values_normalized = [text_to_num(income[0]) if len(income) > 0 else None, text_to_num(expectancy[0]) if len(expectancy) > 0 else None]
        ax.scatter(values_normalized[0], values_normalized[1], c="royalblue")
        
year = 1900
countries = load("./income_per_person_gdppercapita_ppp_inflation_adjusted.csv")["country"].values
plot_income_against_life_expectancy(countries, year)
plt.title(f"Year {year}")
plt.xlabel('Gross Domestic Product')
plt.ylabel('Life Expectancy')
plt.show()
````
