````python
import pandas

def load(path: str) -> pandas.core.frame.DataFrame:
    try:
        return pandas.read_csv(path)
    except FileNotFoundError as msg:
        print(msg)
    return pandas.core.frame.DataFrame()

print(load("./life_expectancy_years.csv"))
````
